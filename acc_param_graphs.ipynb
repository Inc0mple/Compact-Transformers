### Accuracy vs parameter count graphs

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick
```

```python
def plot_accparam_graph(data, yticks):
    fig, ax = plt.subplots(figsize=(8,8))

    for model_class in data.keys():
        ax.plot(data[model_class]['x'], data[model_class]['y'], marker='o', label=model_class)
    
    ax.xaxis.set_major_formatter(mtick.ScalarFormatter(useOffset=False, useMathText=True))
    ax.set_xticks([i*10**6 for i in range(1, 5)])
    
    plt.legend()
    plt.yticks(np.arange(yticks[0], yticks[1], yticks[2]))
    plt.xlabel('No. of parameters')
    plt.ylabel('Top-1 evaluation accuracy (%)')
    plt.title('Accuracy vs parameter count (CIFAR-10)')
    plt.show()
```

```python
# CIFAR-10
c10_data = {"CCT": {'x': [283723, 3233803, 3760139], 'y': [87.23, 95.38, 94.61]}, 
        # "CCT_4xEnsemble": {'x': [1136942], 'y': [89.24]}, 
        'CCT_DynEmbed': {'x': [316747, 3365387, 3891723], 'y': [87.95, 95.47, 95.65]}, 
        'CCT_DynEmbedScaledAttn': {'x': [316749, 3365393, 3891730], 'y': [88.18, 95.49, 95.66]}, 
        # 'CCT_DynEmbedScaledAttnFactor2': {'x': [3169553, 3663250], 'y': [93.76, 94.47]}
        'CCT_SETok': {'x': [286487, 3242267, 3768603], 'y': [86.9, 95.08, 95.5]}
       }

plot_accparam_graph(c10_data, (87, 97, 1))
```

```python
# CIFAR-100
c100_data = {"CCT": {'x': [295333, 3256933, 3783269], 'y': [62.78, 80.27, 80.99]}, 
        'CCT_DynEmbed': {'x': [328357, 3388517, 3914853], 'y': [65.59, 79.69, 81.19]}, 
        'CCT_DynEmbedScaledAttn': {'x': [328359, 3388523, 3914860], 'y': [65.67, 80.55, 81.19]}, 
        'CCT_SETok': {'x': [298097, 3265397, 3791733], 'y': [62.15, 80.29, 80.54]}
       }

plot_accparam_graph(c100_data, (62.5, 83, 2.5))
```
